```python
import pandas as pd

# Load your dataset into a pandas dataframe (replace 'data.csv' with your actual filename)
df = pd.read_csv('/content/Dataset .csv')

# Split the 'Cuisines' column into a list for each row
df['Cuisines'] = df['Cuisines'].str.split(',')

# Explode the dataframe to create separate rows for each cuisine offered by a restaurant
df = df.explode('Cuisines')

# Get the count of each unique cuisine
cuisine_counts = df['Cuisines'].value_counts()

# Sort the cuisine counts by descending order
cuisine_counts = cuisine_counts.sort_values(ascending=False)

# Get the top 3 cuisines
top_3_cuisines = cuisine_counts.head(3)

# Calculate the percentage of restaurants serving each top cuisine
total_restaurants = len(df)
for cuisine, count in top_3_cuisines.items():
  percentage = (count / total_restaurants) * 100
  print(f"{cuisine}: {percentage:.2f}%")
```

```python
import pandas as pd

# Load your dataset into a pandas dataframe (replace 'data.csv' with your actual filename)
df = pd.read_csv('/content/Dataset .csv')

# Split the 'Cuisines' column into a list for each row, handling potential non-string values
df['Cuisines'] = df['Cuisines'].apply(lambda x: x.split(',') if isinstance(x, str) else [])

# Explode the dataframe to create separate rows for each cuisine offered by a restaurant
df_exploded = df.explode('Cuisines')

# Create a new column with cuisine combinations (sorted for order-independence)
df_exploded['Cuisine Combination'] = df_exploded['Cuisines'].apply(lambda x: sorted(x) if isinstance(x, list) else [])

# Count occurrences of each cuisine combination
cuisine_combo_counts = df_exploded['Cuisine Combination'].value_counts()

# Sort cuisine combinations by count (descending)
cuisine_combo_counts = cuisine_combo_counts.sort_values(ascending=False)

# Get the top N cuisine combinations (adjust N as needed)
N = 10  # Replace with the desired number of top combinations
top_n_combos = cuisine_combo_counts.head(N)

# Group data by cuisine combination and calculate average rating (modify if 'Rating' is in a different column)
avg_rating_by_combo = df_exploded.groupby('Cuisines')['Aggregate rating'].mean()

# Print results
print("Top Cuisine Combinations and Average Ratings:")
for combo, count in top_n_combos.items():
  avg_rating = avg_rating_by_combo.get(combo)  # Handle potential missing combinations
  if avg_rating is not None:
    # Check if avg_rating is a single value or a Series
    if isinstance(avg_rating, pd.Series):
      avg_rating = avg_rating.iloc[0] if not avg_rating.empty else None  # Extract the value if it's a non-empty Series

    if avg_rating is not None:
      print(f"{combo}: {count} occurrences (average rating: {avg_rating:.2f})")
    else:
      print(f"{combo}: {count} occurrences (average rating: data not available)")
  else:
    print(f"{combo}: {count} occurrences (average rating: data not available)")
```

```python
import pandas as pd

# Load your dataset into a pandas dataframe (replace 'data.csv' with your actual filename)
df = pd.read_csv('/content/Dataset .csv')

# Identify rows with potential chain names (assuming 'Name' column contains restaurant names)
potential_chains = df[df['Restaurant Name'].str.contains(' ', regex=True)]

# Function to check for potential chain indicators in the name
def is_chain(name):
  # Replace with your logic to identify chain indicators in the name (e.g., "McDonalds", "KFC", "Dominos")
  indicators = ["McDonalds", "KFC", "Dominos", "Starbucks"]  # Example indicators (modify as needed)
  for indicator in indicators:
    if indicator.lower() in name.lower():
      return True
  return False

# Filter for potential chains based on the function
chains = potential_chains[potential_chains['Restaurant Name'].apply(is_chain)]

# Calculate total restaurants
total_restaurants = len(df)

# Analyze chains (if any are found)
if len(chains) > 0:
  # Group data by chain name and calculate average rating and number of restaurants
  chain_data = chains.groupby('Restaurant Name').agg({'Rating': 'mean', 'Name': 'count'})
  chain_data.columns = ['Aggregate Rating', 'Number of Restaurants']

  # Sort by number of restaurants (descending)
  chain_data = chain_data.sort_values(by='Number of Restaurants', ascending=False)

  print("Restaurant Chain Analysis:")
  print(chain_data)
else:
  print("No restaurant chains found in the dataset.")
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')

import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# Load your dataset into a pandas dataframe (replace 'data.csv' with your actual filename)
df = pd.read_csv('/content/Dataset .csv')

# Assuming 'Review' is the column containing review text
reviews = df['Rating text'].tolist()

# Function to preprocess text reviews (remove stopwords, lowercase)
def preprocess_review(text):
  stop_words = stopwords.words('english')
  tokens = word_tokenize(text)
  filtered_tokens = [token.lower() for token in tokens if token not in stop_words]
  return ' '.join(filtered_tokens)

# Preprocess all reviews
preprocessed_reviews = [preprocess_review(review) for review in reviews]

# Combine reviews into a single string for easier analysis (optional)
all_text = ' '.join(preprocessed_reviews)

# Use a sentiment analysis library (e.g., TextBlob, VADER) to identify positive and negative words

from textblob import TextBlob

blob = TextBlob(all_text) # Create a TextBlob object

# Find most common positive and negative words (adjust number as needed)
# Iterate through words in the blob and check their sentiment polarity
positive_words = [word for word in blob.words if blob.sentiment.polarity > 0]
negative_words = [word for word in blob.words if blob.sentiment.polarity < 0]

from collections import Counter

# Count the frequency of positive and negative words
positive_word_counts = Counter(positive_words).most_common(20)
negative_word_counts = Counter(negative_words).most_common(20)


# Calculate average review length
average_length = sum(len(review) for review in reviews) / len(reviews)

# Explore relationship between review length and rating (optional)
# You can use correlation analysis (e.g., df['Rating'].corr(df['Review'].str.len()))
# or group data by review length and calculate average rating per group

print("Most Common Positive Words:")
print(positive_word_counts) # Print the most common positive words and their counts

print("\nMost Common Negative Words:")
print(negative_word_counts) # Print the most common negative words and their counts

print(f"\nAverage Review Length: {average_length:.2f} characters")

# Add your analysis of the relationship between review length and rating here (optional)
```

```python
import pandas as pd

# Load your dataset into a pandas dataframe (replace 'data.csv' with your actual filename)
df = pd.read_csv('/content/Dataset .csv')

# Assuming 'Price Range', 'Online Delivery' (Yes/No), and 'Table Booking' (Yes/No) are column names
price_ranges = df['Price range'].unique()  # Get unique price ranges

# Create a dictionary to store counts for each combination
availability_counts = {}
for price_range in price_ranges:
  availability_counts[price_range] = {
      'Delivery_Yes': 0, 'Delivery_No': 0,
      'Booking_Yes': 0, 'Booking_No': 0,
  }

# Count occurrences based on price range, delivery, and booking availability
for index, row in df.iterrows():
  price_range = row['Price range']
  delivery = row['Has Online delivery']
  booking = row['Has Table booking']

  availability_counts[price_range][f'Delivery_{delivery}'] += 1
  availability_counts[price_range][f'Booking_{booking}'] += 1

# Print results (modify for readability as needed)
print("Availability by Price Range:")
for price_range, counts in availability_counts.items():
  print(f"\nPrice range: {price_range}")
  print(f"  Delivery (Yes): {counts['Delivery_Yes']}")
  print(f"  Delivery (No): {counts['Delivery_No']}")
  print(f"  Table Booking (Yes): {counts['Booking_Yes']}")
  print(f"  Table Booking (No): {counts['Booking_No']}")

# (Optional) Calculate percentages for better comparison
# You can modify the printing section to calculate and display percentages
# for each combination within a price range.
```
